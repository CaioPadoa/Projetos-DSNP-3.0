# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

O seguinte dataset foi obtido através do repositório da [Our World in Data](https://github.com/owid/covid-19-data/tree/master/public/data//), no GitHub.

Todos os dados foram coletados pelo "System Science and Engineering" da Universidade de Johns Hopkins. Eles são coletados constantemente, e muitas das entradas são atualizadas diariamente.

Os dados desta análise foram coletados até dia 17/06/2022. 

Não é surpresa para ninguém que a coleta de dados relacionados ao COVID-19 é um desafio para todas as instituições governamentais e não-governamentais desde o início da pandemia.

Isso se dá por diversos motivos, desde a falta de infraestrutura até a arbitrária decisão de não divulgar dados publicamente - muitas vezes por motivos políticos.

Entretanto, este conjunto de dados é o mais próximo que podemos chegar da realidade. Coletados por uma instituição de renome, é possível afirmar com certa garantia que os números deste dataset podem nos ajudar a tirar diversas conclusões interessantes para entender como ocorreu a pandemia no Brasil e no resto do mundo.

Como dito anteriormente, diversas colunas são compostas por muitos valores ausentes. Entretanto, tendo em vista este dataset, isso não necessariamente é um problema.

Em algumas colunas, um valor ausente pode significar simplesmente que não houve nenhuma entrada neste dia. Explico com um exemplo:

Se determinada data possui o valor de "total_deaths" ausente, significa que não houve nenhuma morte neste dia. 

Por este motivo, troquei todos os valores ausentes deste dataset por 0 com o código acima.

```
#ver quais países têm maior número de casos
df.loc[df.date == "2022-06-16"].sort_values(by="total_cases", ascending=False)
```

Esse filtro nos mostra alguns dados interessantes:
* A Ásia foi o continente que mais relatou casos de COVID.
* O Brasil foi, de longe, o país que mais sofreu com casos de COVID, com 31 milhões de casos até a data da última atualização deste dataset (04/06/2022).

```
#ver quais países têm maior número de mortes
df.loc[df.date == "2022-06-16"].sort_values(by="total_deaths", ascending=False)
```

Novamente, o Brasil lidera o ranking de países com maior número de mortos por COVID, com cerca de 666 mil falecidos até a última coleta de dados.

Como esperado, o ranking de países com maior número de mortos é extremamente semelhante ao ranking de casos. Isso indica que nenhum país teve uma taxa de mortalidade extremamente elevada em relação às demais. 

```
fig, ax = plt.subplots(figsize = (10,5))

top5_total_deaths = df.loc[df.date == "2022-06-16", ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]
ax.barh('location', 'total_deaths', data=top5_total_deaths, color = "royalblue", edgecolor = "black", )
plt.title("Total de mortes por região")

plt.tight_layout();
```

Este gráfico mostra a comparação entre o número de mortos por região.

Analisando desta maneira, é visível o quão calamitosa foi a pandemia no Brasil, tendo em vista que o total de óbitos no país é consideravelmente maior do que o de outros países do ranking.

```
#plotando gráfico de dispersão entre PIB per capita e total de mortos
fig, ax = plt.subplots(figsize = (10, 5))

df.plot.scatter(x = "gdp_per_capita", y = "total_deaths", ax = ax)
ax.set_title("PIB per capita e Total de Mortos")
```

Analisando um gráfico de dispersão entre o PIB per capita dos países registrados no dataset e o número de mortos por COVID nestes países, é possível visualizar claramente que as nações com menor PIB per capita sofreram com mais mortes do que aquelas com alto desenvolvimento econômico.

Essa tese, a princípio, faz sentido, tendo em vista que essas nações não teriam as mesmas condições de cuidar de seus infectados que países mais desenvolvidos teriam.

Entretanto, é necessário fazer uma rápida análise de correlação para entender se isso realmente procede.

```
corr_pib_mortos = df[["total_deaths", "gdp_per_capita"]].corr()

sns.heatmap(corr_pib_mortos, annot = True, cmap = "Blues")

plt.show()
```

Como visto no gráfico acima, a correlação entre o PIB per capita e o número total de mortos dos países é fraca e negativa.

Isso significa que essa hipótese, muito provavelmente, não é verdadeira.

Entretanto, para poder afirmar com mais convicção, seria necessário fazer um modelo de regressão linear, e diversos testes estatísticos, para descobrir o quanto da variável "total_deaths" é explicado pela variável "gdp_per_capita".

### Análise Exploratória para o Brasil

Após visualizar rapidamente como anda o cenário do COVID no mundo, hora de entender como a doença se comportou no Brasil.

Para isso, filtraremos nosso dataset, e analisaremos apenas os números da nossa nação.

```
df_brasil = df.loc[df.location == "Brazil"].copy()

display(df_brasil.head())
display(df_brasil.tail())
```

```
df_brasil.fillna(0,inplace = True)
df_brasil.head()
```

Aqui, foi feito o mesmo que no início desta análise. Substituí os valores ausentes por 0.

```
#data da primeira morte no Brasil
for value in df_brasil["total_deaths"]:
  if value != 0:
    primeira_morte = (df_brasil.loc[df_brasil["total_deaths"] == value])
    display(primeira_morte)
    print("\nA data em que ocorreu a primeira morte por COVID no Brasil é: {}".format(primeira_morte.date.iloc[0]))
    break
```

```
for value in df_brasil["total_cases"]:
  if value != 0:
    primeiro_caso = (df_brasil.loc[df_brasil["total_cases"] == value])
    display(primeiro_caso)
    print("\nO primeiro caso foi detectado em: {}".format(primeiro_caso.date.iloc[0]))
    break
```

```
primeira_morte.date.iloc[0] - primeiro_caso.date.iloc[0] #deixar mais bonito
```

Nesta parte da análise, entendemos quando ocorreu o primeiro caso e a primeira morte no Brasil.

Com isso, foi possível descobrir quantos dias se passaram entre a primeira testagem positiva e o primeiro óbito por COVID-19 no país. E a resposta é: 20 dias.

```
#visualizando o histório de testagens positivas para COVID-19 no Brasil ao longo do tempo
df_brasil_date = df_brasil.set_index("date").copy()

fig, (ax1, ax2) = plt.subplots(1,2, figsize = (10,5))

df_brasil_date["total_cases"].plot(kind = "line", ax=ax1)
ax1.set_title("Total de casos de COVID no Brasil (escala linear)")
ax1.set_ylabel("Casos (em milhão)") 
ax1.set_xlabel("Data")

df_brasil_date["total_cases"].plot(kind = "line", ax=ax2)
ax2.set_title("Total de casos de COVID no Brasil (escala logarítmica)")
ax2.set_ylabel("Casos") 
ax2.set_xlabel("Data")
ax2.set_yscale("log")

plt.tight_layout();
```

Neste gráfico, é possível visualizar o rápid crescimento do número de casos no Brasil ao longo do tempo, em escala liner e logarítimica.

Essa visualização comprova que, até mentade de 2021, o crescimento era praticamente exponencial.

Entretanto, com as medidas restritivas, vacinação e maior imunidade da população, a curva foi achatada durante alguns meses, voltando a subir no final de 2021 e início de 2022.

Até a última coleta de dados do dataset (metade de 2022), a curva está se achatando novamente.

```
fig, (ax3, ax4) = plt.subplots(1,2, figsize = (10,5))

df_brasil_date["total_deaths"].plot(kind = "line", ax=ax3)
ax3.set_title("Total de mortes por COVID no Brasil (escala linear)")
ax3.set_ylabel("Mortes") #como mudar a escala?
ax3.set_xlabel("Data")

df_brasil_date["total_deaths"].plot(kind = "line", ax=ax4)
ax4.set_title("Total de mortes por COVID no Brasil (escala logarítmica)")
ax4.set_ylabel("Mortes") #como mudar a escala?
ax4.set_xlabel("Data")
ax4.set_yscale("log")

plt.tight_layout()
```

O gráfico de mortes por COVID no Brasil ao longo do tempo apresenta um comportamente muito semelhante ao de testagens positivas.

```
#Brasil comparado com resto do mundo
print("A taxa de mortalidade do COVID no Brasil é de: {:.3}%".format((df_brasil["total_deaths"].sum()/df_brasil["total_cases"].sum())*100))
print("\nA taxa de mortalidade do COVID no mundo é de: {:.3}%\n".format((df["total_deaths"].sum()/df["total_cases"].sum())*100))

fig, (ax5, ax6) = plt.subplots(1, 2, figsize = (10,5))

df_brasil_date["total_cases"].plot(kind = "line", ax = ax5)
ax5.set_title("Avanço no número de casos no Brasil")
ax5.set_ylabel("Casos (em milhões)")


df_date = df.set_index("date").copy()
df_date["total_cases"].plot(kind = "line", ax = ax6)
ax6.set_title("Avanço no número de casos no mundo")
ax6.set_ylabel("Casos (em dezenas de milhões)")

plt.tight_layout();
```

Este gráfico mostra como o avanço no número de casos no Brasil e no mundo foi bem similar.
